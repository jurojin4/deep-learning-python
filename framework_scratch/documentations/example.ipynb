# **2D Example**

```python
from sklearn.datasets import make_blobs

import numpy as np

X, Y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=0)

# Normalization
X_max = X.max()
X_min = X.min()
for i in range(X.shape[1]):
    X[:, i] = (X[:, i] - X_min) / (X_max - X_min)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

fig, ax = plt.subplots(1, 2, figsize=(24, 8))

ax[0].scatter(X[:, 0], X[:, 1])

ax[0].set_xlim(-0.5, 1.25)
ax[0].set_ylim(-0.25, 1.25)
ax[0].spines['bottom'].set_position(('data', 0))
ax[0].spines['left'].set_position(('data', 0))
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[0].set_title("Affichage sans classes\n", fontsize=24)

sc = ax[1].scatter(X[:, 0], X[:, 1], c=Y, cmap="rainbow")

ax[1].set_xlim(-0.5, 1.25)
ax[1].set_ylim(-0.25, 1.25)
ax[1].spines['bottom'].set_position(('data', 0))
ax[1].spines['left'].set_position(('data', 0))
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
ax[1].set_title("Affichage avec classes\n", fontsize=24)

classes = np.unique(Y)
handles = [plt.Line2D([], [], marker="o", ls="", color=sc.cmap(sc.norm(c)))
           for c in classes]
ax[1].legend(handles, [f"Classe {c}" for c in classes],
             fontsize=12, title_fontsize=14)

plt.savefig(os.path.join(plots_path, "2d_example"), pad_inches=0.)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

fig, ax = plt.subplots(1, 2, figsize=(24, 8))

x = np.linspace(-1, 1, X.shape[0])

y = -(2.79*x + 4) / -9.38

ax[0].scatter(X[:, 0], X[:, 1])
ax[0].plot(x, y, c='green', lw=3, label="Z")

ax[0].set_xlim(-0.5, 1.25)
ax[0].set_ylim(-0.25, 1.25)
ax[0].spines['bottom'].set_position(('data', 0))
ax[0].spines['left'].set_position(('data', 0))
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[0].set_title("Affichage sans classes\n", fontsize=24)

ax[0].legend(loc="upper right")

sc = ax[1].scatter(X[:, 0], X[:, 1], c=Y, cmap="rainbow")
ax[1].plot(x, y, c='green', lw=3, label="Z")

ax[1].set_xlim(-0.5, 1.25)
ax[1].set_ylim(-0.25, 1.25)
ax[1].spines['bottom'].set_position(('data', 0))
ax[1].spines['left'].set_position(('data', 0))
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
ax[1].set_title("Affichage avec classes\n", fontsize=24)

classes = np.unique(Y)
handles = [plt.Line2D([], [], marker="o", ls="", color=sc.cmap(sc.norm(c))) for c in classes]
labels = [f"Classe {c}" for c in classes]

line_z, = ax[1].plot([], [], c='green', lw=3, label="Z")

handles.append(line_z)
labels.append("Z")

ax[1].legend(handles, labels, fontsize=12, title_fontsize=14)

plt.savefig(os.path.join(plots_path, "2d_example_with_linear_separation"), pad_inches=0.)
```
